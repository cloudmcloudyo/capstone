```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.imsdb.com/TV/Seinfeld.html'
res = requests.get(url)
```

```python
res.status_code
```

```python
soup = BeautifulSoup(res.content, 'lxml')
```

```python
link = soup.find_all('a')
```

```python
link
```

```python
link = link[64:-7]
link
```

```python
link[0].text
```

```python
branch = link[0].attrs['href']
branch
```

```python
branch_1 = branch[4:24]
branch_1 = branch_1.replace('T', 't')
```

```python
branch_2 = branch[27:-12]
branch_2 = branch_2.replace(' ', '-')
```

```python
branch_new = branch_1 + '-' + branch_2 + '.html'
```

```python
url = url[:-16]
url
```

```python
ep_url = url + branch_new
```

```python
ep_res = requests.get(ep_url)
```

```python
ep_res.status_code
```

```python
ep_soup = BeautifulSoup(ep_res.content, 'lxml')
```

```python
script = ep_soup.find('td', {'class': 'scrtext'})
script
```

```python
script.find_all('pre')[0].text
```

```python
text = script.find_all('pre')[0].text.strip().split('\n')
text
```

```python
# while loop start token: ALL CAPS (if .upper() ==) , then appending, stop when ()(regex) or all caps
```

```python
t1= [x.strip() for x in text if x.strip()]
t1
```

```python
list(enumerate(t1))
```

```python
def replace_space(text):
    res = ''
    for i in text:
        res += i.replace('               ', '')
    return res
```

```python
combined_text = replace_space(text)
combined_text
```
